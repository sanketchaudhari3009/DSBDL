```python
import docx2txt
import spacy
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
```

```python
text=docx2txt.process('testdoc.docx')
text2=docx2txt.process('testdoc2.docx')
```

```python
nlp=spacy.load('en_core_web_md')
```

```python
d1=nlp(text)
d2=nlp(text2)
```

## Tokenization

```python
tokanized_d1=[]
for token in d1:
  tokanized_d1.append(token)
tokanized_d1
```

```python
tokanized_d2=[]
for token in d2:
  tokanized_d2.append(token)
tokanized_d2
```

## POS Tagging

```python
for token in d1:
  print(token,"->>>>>",token.pos_)
```

```python
for token in d2:
  print(token,"->>>>>",token.pos_)
```

## Stop Word Removal

```python
nltk.download('stopwords')

wosd1=[]
for word in d1:
  if(word.text not in stopwords.words()):
    wosd1.append(word.text);
print(len(wosd1)," < ",len(d1))
```

```python
wosd2=[]
for word in d2:
  if(word.text not in stopwords.words()):
    wosd2.append(word.text);
print(len(wosd2)," < ",len(d2))
```

## Stemming

```python
ps=PorterStemmer()
stemmed_d1=[]
for word in wosd1:
  stemmed_d1.append(ps.stem(word))
stemmed_d1
```

```python
ps=PorterStemmer()
stemmed_d2=[]
for word in wosd2:
  stemmed_d2.append(ps.stem(word))
stemmed_d2
```

## Lemmatizaiton

```python
nltk.download('wordnet')

lemmatized_d1=[]
wl=WordNetLemmatizer()
for word in wosd1:
  lemmatized_d1.append(wl.lemmatize(word))
lemmatized_d1
```

```python
lemmatized_d2=[]
for word in wosd2:
  lemmatized_d2.append(wl.lemmatize(word))
lemmatized_d2
```

## TF-IDF

```python
#calculated frequency of every word
tf=[]
corpus=[lemmatized_d1,lemmatized_d2]
for data in corpus:
  count={}
  for word in data:
    if word in count:
      count[word]+=1
    else:
      count[word]=1
  tf.append(count)
tf
```

```python
#dividing the occurence with length
for i in range(len(corpus)):
  length=len(corpus[i])
  for words in tf[i].keys():
    tf[i][words]=tf[i][words]/length
  
```

```python
#calculated frequency for IDF

corpus=[set(lemmatized_d1),set(lemmatized_d2)]
unique_word=set(lemmatized_d1).union(set(lemmatized_d2))
idf={} 
for word in unique_word:
  idf[word]=0
  for data in corpus:
    if word in data:
      idf[word]+=1
idf
```

```python
#calculate idf
length_of_corpus=len(corpus)
for word in idf.keys():
  idf[word]=log10(length_of_corpus/idf[word])
```

```python
#tf*idf
tf_idf=[]
for sentence in tf:
  dict_sent={}
  for words in sentence.keys():
    dict_sent[words]=sentence[words]*idf[words]
  tf_idf.append(dict_sent)
tf_idf
```
