```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("nba.csv")
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Weight']=df['Weight'].fillna(df['Weight'].mean())
```

```python
df['Number']=df['Number'].fillna(df['Number'].mode())
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].quantile(0.5)) #similar to median()
```

```python
#data transformation scaling for change
#minmax
numeric=df.select_dtypes(include=np.number)
```

```python
for i in numeric.columns:
    numeric[i]=(numeric[i]-numeric[i].min())/(numeric[i].max()-numeric[i].min())
numeric
```

```python
numeric.plot(kind='bar')
```

```python
numeric.plot(kind='hist')
```

```python
numeric.plot(kind='box')
```

```python
zs=df.select_dtypes(include=np.number)
```

```python
def normality(data, feature):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(data[feature])
    plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()
normality(df,'Weight2')
```
