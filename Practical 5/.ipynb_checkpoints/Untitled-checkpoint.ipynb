```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import preprocesing
```

```python
df=pd.read_csv('nba.csv')
```

```python
df.dtypes
```

```python
print("Mean of age: ", df['Age'].mean())
print("Median of age: ", df['Age'].median())
print("Min of age: ", df['Age'].min())
print("Max of age: ", df['Age'].max())
print("SD of age: ", df['Age'].std())
```

```python
print("Mean of salary: ", df['Salary'].mean())
print("Median of salary: ", df['Salary'].median())
print("Min of salary: ", df['Salary'].min())
print("Max of salary: ", df['Salary'].max())
print("SD of salary: ", df['Salary'].std())
```

```python
bins=[19,25,31,36,40]
labels=['19-24','25-30','31-35','36-40']
df['AgeGroup']=pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df
```

```python
df.groupby('AgeGroup')['Salary'].mean()
df.groupby('AgeGroup')['Salary'].mean().describe()
```

```python
salaryList=list(df.groupby('AgeGroup')['Salary'])
salaryList
```

```python
pivot_table = df.pivot_table(values='Age', index='Salary', columns=df['AgeGroup'], aggfunc=['mean', 'median', 'min', 'max', 'std'])
pivot_table
```

```python
crosstab = pd.crosstab(index=df['Salary'], values=df['Salary'], columns=df['AgeGroup'], aggfunc=['mean', 'median', 'min', 'max', 'std'])
crosstab
```
